```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
books=[]
for i in range(1,51):
    url = f'https://books.toscrape.com/catalogue/page-{i}.html'
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response, 'html.parser')
    ol = soup.find('ol')
    articles = ol.find_all('article',class_='product_pod')
    


    for article in articles:
        image = article.find('img')
        title = image.attrs['alt']
        star = article.find('p')
        star= star['class'][1]
        price = article.find('p',class_='price_color').text
        price = float(price[1:])
        
        # Get detail page URL
        relative_link = article.find('h3').find('a')['href']
        detail_url = 'https://books.toscrape.com/catalogue/' + relative_link.replace('../../../', '')

        # Visit detail page to get category
        detail_response = requests.get(detail_url)
        detail_soup = BeautifulSoup(detail_response.content, 'html.parser')

        category = detail_soup.find('ul', class_='breadcrumb').find_all('li')[2].a.text
        availability = detail_soup.find('p', class_='instock availability').get_text(strip=True)
        books.append([title,price,star,category,availability])

    
```

```python
df = pd.DataFrame(books, columns= ['Title','Price','Star','category','availability'])
```
